```python
import pandas as pd
```

```python
df = pd.read_csv('data/starbucks.csv', index_col='Date', parse_dates=True)
df.head()
```

```python
df['Close'].plot(figsize=(12,5))
```

Rolling Mean: taking average of the data based on given window. for example window=5 then it will discard first 4 values and 5th value will be average of all 5 values and so on.

```python
df['Close']['2015-01-01':'2015-12-31'].plot(figsize=(12,5))
df.rolling(window=30).mean()['Close']['2015-01-01':'2015-12-31'].plot()
```

```python
df['Close: 5 Day Mean'] = df['Close'].rolling(window=5).mean()
df.head(10)
```

```python
df[['Close', 'Close: 5 Day Mean']].plot(figsize=(12,5))
```

with legend

## Expanding

Instead of calculating values for a rolling window of dates, what if you wanted to take into account everything from the start of the time series up to each point in time? For example, instead of considering the average over the last 7 days, we would consider all prior data in our expanding set of averages.

```python
df['Close'].plot()
df['Close'].expanding(min_periods=1).mean().plot(figsize=(12,5))
```
